```python
import requests
from bs4 import BeautifulSoup
#Creating Header

headers = {
    "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36\", \"(KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36",
    "Referer": "https://www.google.com/search?q=valor&oq=valor+&aqs=chrome..69i57j69i64j35i39i650j35i39j69i60l4.2669j0j1&sourceid=chrome&ie=UTF-8",
    "Accept-Language": "pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7"
}
```

```python
#URL, Request and bs4 parser

URL = "https://valor.globo.com/"

r = requests.get(URL, headers=headers)

soup = BeautifulSoup(r.content, "html.parser")
```

```python
headlinesPublic = soup.find_all("a", {'class': 'bstn-dedupe-url'})
headlinesPaid = soup.find_all("a", {'class': 'bstn-dedupe-url is-subscriber-only'})

# Creating a list of headlines and their corresponding href
headlines_publicList = []
hyperlink_publicList = []

for headline in headlinesPublic:
    headlines_publicList.append(headline.text)
    hyperlink_publicList.append(headline['href'])

headlines_paidList = []
hyperlink_paidList = []

for headline in headlinesPaid:
    headlines_paidList.append(headline.text)
    hyperlink_paidList.append(headline['href'])

# Updating the headlines_publicList to exclude lines starting with "— Em"
headlines_publicList_updated = []
hyperlink_publicList_updated = []

for i in range(len(headlines_publicList)):
    if not headlines_publicList[i].strip().startswith("— Em"):
        headlines_publicList_updated.append(headlines_publicList[i])
        hyperlink_publicList_updated.append(hyperlink_publicList[i])

# Updating the headlines_paidList to exclude lines starting with "— Em"
headlines_paidList_updated = []
hyperlink_paidList_updated = []

for i in range(len(headlines_paidList)):
    if not headlines_paidList[i].strip().startswith("— Em"):
        headlines_paidList_updated.append(headlines_paidList[i])
        hyperlink_paidList_updated.append(hyperlink_paidList[i])

# Printing the updated lists
for i in range(len(headlines_publicList_updated)):
    print(headlines_publicList_updated[i], ":", hyperlink_publicList_updated[i])

for i in range(len(headlines_paidList_updated)):
    print(headlines_paidList_updated[i], ":", hyperlink_paidList_updated[i])
```
